```python
xy = np.array([[828.659973, 833.450012, 908100, 828.349976, 831.659973],
 [823.02002, 828.070007, 1828100, 821.655029, 828.070007],
 [819.929993, 824.400024, 1438100, 818.97998, 824.159973],
 [816, 820.958984, 1008100, 815.48999, 819.23999],
 [819.359985, 823, 1188100, 818.469971, 818.97998],
 [819, 823, 1198100, 816, 820.450012],
 [811.700012, 815.25, 1098100, 809.780029, 813.669983],
 [809.51001, 816.659973, 1398100, 804.539978, 809.559998]])
```

```python
print(xy[0:3])
```

```python
x_data = xy[:, 0:-1]
```

```python
print(x_data[0:3])
```

```python
y_data = xy[:, [-1]]
```

```python
print(y_data[0:3])
```

```python
X = tf.placeholder(shape=[None, 4], dtype=tf.float32, name='X')
```

```python
Y = tf.placeholder(shape=[None, 1], dtype=tf.float32, name='Y')
```

```python
W = tf.Variable(tf.random_normal(shape=[4,1]), shape=[4,1], dtype=tf.float32, name='weight')
```

```python
b = tf.Variable(tf.random_normal(shape=[1]), shape=[1], dtype=tf.float32, name='bias')
```

```python
hypothesis = tf.matmul(X, W) + b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(2001):
        cost_val, W_val, b_val, predicted_Y, observed_Y, _ = sess.run([cost, W, b, hypothesis, Y, optimizer], feed_dict={X: x_data, Y:y_data})
        if step%100 == 0:
            print(step, cost_val, W_val, b_val, predicted_Y, observed_Y)
```

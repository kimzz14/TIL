```python
import tensorflow as tf
```

```python
learning_rate = 0.01
```

```python
x_data = [1, 2, 3]
```

```python
y_data = [1, 2, 3]
```

```python
X = tf.placeholder(shape=[3], dtype=tf.float32, name='weight')
```

```python
Y = tf.placeholder(shape=[3], dtype=tf.float32, name='bias')
```

```python
W = tf.Variable(tf.random_normal(shape=[1]), shape=[1], dtype=tf.float32, name='weight')
```

```python
hypothesis = X * W
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
gradient = 2 * tf.reduce_mean((X * W - Y) * X)
```

```python
train = W.assign(W - learning_rate * gradient)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(2001):
        cost_val, W_val, _ = sess.run([cost, W, train], feed_dict={X: x_data, Y: y_data})
        if step% 20 == 0:
            print(step, cost_val, W_val)
        
        
```

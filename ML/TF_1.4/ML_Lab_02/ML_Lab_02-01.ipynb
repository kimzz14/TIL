```python
import tensorflow as tf
```

```python
x_train = [1, 2, 3]
```

```python
y_train = [1, 2, 3]
```

```python
X = tf.placeholder(shape=[3], dtype=tf.float32, name='X')
```

```python
Y = tf.placeholder(shape=[3], dtype=tf.float32, name='Y')
```

```python
W = tf.Variable(tf.random_normal(shape=[1]), shape=[1], name='weight')
```

```python
b = tf.Variable(tf.random_normal(shape=[1]), shape=[1], name='bias')
```

```python
hypothesis = X * W + b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
```

```python
train = optimizer.minimize(cost)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(2001):
        cost_val, W_val, b_val, _ = sess.run([cost, W, b, train], feed_dict={X: x_train, Y: y_train})
        if step% 20 == 0:
            print(step, cost_val, W_val, b_val)
```

```python
import tensorflow as tf
```

```python
import numpy as np
```

```python
x_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32)
```

```python
y_data = np.array([[0], [1], [1], [0]], dtype=np.float32)
```

```python
X = tf.placeholder(shape=[None, 2], dtype=tf.float32, name='X')
```

```python
Y = tf.placeholder(shape=[None, 1], dtype=tf.float32, name='Y')
```

```python
W1 = tf.Variable(tf.random_normal(shape=[2, 2]), shape=[2, 2], dtype=tf.float32, name='weight1')
```

```python
b1 = tf.Variable(tf.random_normal(shape=[2]), shape=[2], dtype=tf.float32, name='bias1')
```

```python
layer1 = tf.sigmoid(tf.matmul(X, W1) + b1)
```

```python
W2 = tf.Variable(tf.random_normal(shape=[2, 1]), shape=[2, 1], dtype=tf.float32, name='weight2')
```

```python
b2 = tf.Variable(tf.random_normal(shape=[1]), shape=[1], dtype=tf.float32, name='bias2')
```

```python
layer2 = tf.sigmoid(tf.matmul(layer1, W2) + b2)
```

```python
cost = -tf.reduce_mean(Y * tf.log(layer2) + (1 - Y) * tf.log(1 - layer2))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(cost)
```

```python
predicted_Y = tf.cast(layer2 > 0.5, dtype=tf.float32)
```

```python
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted_Y, Y), dtype=tf.float32))
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(10001):
        sess.run(optimizer, feed_dict={X: x_data, Y: y_data})
        if step%100 == 0:
            print(step, sess.run(cost, feed_dict={X: x_data, Y: y_data}), sess.run([W1, W2]))
    h, c, a = sess.run([layer2, predicted_Y,accuracy], feed_dict={X: x_data, Y: y_data})
    print('\nHypothesis: ', h, '\nCorrect: ', c, '\nAccuracy: ', a)
```

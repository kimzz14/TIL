```python
import tensorflow as tf
```

```python
x_data = tf.Variable([[73,80,75],[93,88,93],[89,91,90],[96,98,100],[73,66,70]], shape=[5,3], dtype=tf.float32)
```

```python
y_data = tf.Variable([[152],[185],[180],[196],[142]], shape=[5,1], dtype=tf.float32)
```

```python
class Model(object):
    def __init__(self):
        self.W = tf.Variable(tf.random.normal(shape=[3,1]), shape=[3,1], dtype=tf.float32, name='Weight')
        self.b = tf.Variable(tf.random.normal(shape=[1])  , shape=[1], dtype=tf.float32, name='bias')
    def __call__(self, X):
        return tf.matmul(X, self.W) + self.b
```

```python
@tf.function
def loss(expected_y, observed_y):
    return tf.reduce_mean(tf.square(expected_y - observed_y))
```

```python
@tf.function
def train(model, inputs, outputs, learning_rate=0.01):
    with tf.GradientTape() as t:
        current_loss = loss(model(inputs), outputs)
    dW, db = t.gradient(current_loss, [model.W, model.b])
    model.W.assign_sub(learning_rate * dW)
    model.b.assign_sub(learning_rate * db)
```

```python
model = Model()
```

```python
import tensorflow as tf
```

```python
#X shape=[5,3]
```

```python
X = tf.Variable([[73,80,75],[93,88,93],[89,91,90],[96,98,100],[73,66,70]], dtype=tf.float32)
```

```python
#Y shape = [5,1]
```

```python
Y = tf.Variable([[152],[185],[180],[196],[142]], dtype=tf.float32)
```

```python
#X [5,3] * W [3,1] = [5,1]
```

```python
W = tf.Variable(tf.random_normal([3,1]), dtype=tf.float32, name='weight')
```

```python
b = tf.Variable(tf.random_normal([1]), dtype=tf.float32, name='bias')
```

```python
hypothesis = tf.matmul(X, W) + b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
```

```python
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
```

```python
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2001):
    cost_val, hy_val, y_val, _ = sess.run([cost, hypothesis, Y, train])
    if step % 20 == 0:
        print(cost_val, hy_val, y_val)
```
